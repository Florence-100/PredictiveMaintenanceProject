```python
#take 1250 datapoints to make 5000 datapoints
state0_data = state0_data[:1250]
state1_data = state1_data[:1250]
state2_data = state2_data[:1250]
state3_data = state3_data[:1250]
```

```python
volt0123 = np.array(state0123_data['Volt'].tolist()).transpose()
rotate0123 = np.array(state0123_data['Rotate'].tolist()).transpose()
pressure0123 = np.array(state0123_data['Pressure'].tolist()).transpose()
vibration0123 = np.array(state0123_data['Vibration'].tolist()).transpose()
```

```python
plt.hist(volt0123, bins=50)
plt.title('Volt for Degradation States')
```

```python
plt.hist(rotate0123, bins=50)
plt.title('Rotate for Degradation States')
```

```python
plt.hist(pressure0123, bins=50)
plt.title('Pressure for Degradation States')
```

```python
plt.hist(vibration0123, bins=50)
plt.title('Vibration for Degradation States')
```

```python
#calculate one standard deviation away from mean

import statistics

voltThreshold1 = statistics.mean(volt0123) + statistics.stdev(volt0123)
rotateThreshold1 = statistics.mean(rotate0123) - statistics.stdev(rotate0123)
pressureThreshold1 = statistics.mean(pressure0123) + statistics.stdev(pressure0123)
vibrationThreshold1 = statistics.mean(vibration0123) + statistics.stdev(vibration0123)

print(f"One standard deviation away from mean for volt is {voltThreshold1}")
print(f"One standard deviation away from mean for rotate is {rotateThreshold1}")
print(f"One standard deviation away from mean for pressure is {pressureThreshold1}")
print(f"One standard deviation away from mean for vibration is {vibrationThreshold1}")
```

```python
#calculate two standard deviation away from mean

voltThreshold2 = statistics.mean(volt0123) + 2*statistics.stdev(volt0123)
rotateThreshold2 = statistics.mean(rotate0123) - 2*statistics.stdev(rotate0123)
pressureThreshold2 = statistics.mean(pressure0123) + 2*statistics.stdev(pressure0123)
vibrationThreshold2 = statistics.mean(vibration0123) + 2*statistics.stdev(vibration0123)

print(f"Two standard deviation away from mean for volt is {voltThreshold2}")
print(f"Two standard deviation away from mean for rotate is {rotateThreshold2}")
print(f"Two standard deviation away from mean for pressure is {pressureThreshold2}")
print(f"Two standard deviation away from mean for vibration is {vibrationThreshold2}")
```

```python
#calculate three standard deviation away from mean

voltThreshold3 = statistics.mean(volt0123) + 3*statistics.stdev(volt0123)
rotateThreshold3 = statistics.mean(rotate0123) - 3*statistics.stdev(rotate0123)
pressureThreshold3 = statistics.mean(pressure0123) + 3*statistics.stdev(pressure0123)
vibrationThreshold3 = statistics.mean(vibration0123) + 3*statistics.stdev(vibration0123)

print(f"Three standard deviation away from mean for volt is {voltThreshold3}")
print(f"Three standard deviation away from mean for rotate is {rotateThreshold3}")
print(f"Three standard deviation away from mean for pressure is {pressureThreshold3}")
print(f"Three standard deviation away from mean for vibration is {vibrationThreshold3}")
```

```python
def percentageAnomaliesIdentified(data, volt_comparator, rotate_comparator, pressure_comparator, vibration_comparator):
    totalCount = len(data)
    anomalyCount = 0
    for reading in data:
        if (reading[0] > volt_comparator) or (reading[1] < rotate_comparator) or (reading[2] > pressure_comparator) or (reading[3] > vibration_comparator):
            #is anomaly
            anomalyCount += 1
    percentage = anomalyCount/totalCount
    return percentage
```

```python
state89_dataArr = state89_data.drop(columns=['Timestep', 'State'])
state89_dataArr = state89_dataArr.to_numpy()
state89_dataArr
```

```python
state89_datalst = state89_dataArr.tolist()
state89_datalst
```

```python
#if 1 standard deviation away
percentageAnomaliesIdentified(state89_datalst, voltThreshold1, rotateThreshold1, pressureThreshold1, vibrationThreshold1)
```
